**Finding 10**<br>
Collaboration with Third-Party Hacking Groups <br>
I-Soon's collaboration with third-party hacking groups might indicate that China is employing an outsourced model for certain espionage activities to maintain plausible deniability. By partnering with independent or semi-affiliated hacking groups, I-Soon could avoid direct attribution for specific cyberattacks. This hypothesis suggests that I-Soon would handle high-level strategy, target selection, and initial reconnaissance, while third-party actors would be contracted to carry out specific operations, such as phishing campaigns, malware deployment, or data exfiltration.

The mention of contractual discussions (like in File 3 regarding the “Xiamen contract”) might reflect logistical coordination between I-Soon and external entities tasked with handling portions of the cyber operations, further supporting the hypothesis that I-Soon operates through proxy groups or contractors. This outsourcing strategy helps mask the direct involvement of the Chinese state in various cyber activities.

**Question Form**<br>
Was I-Soon collaborating with third party hacking groups?

**Validation**<br>
Conversations that reference contracts or exchanges between different entities (e.g., "Xiamen contract") might suggest a business or operational link between I-Soon and other organizations, possibly indicating outsourced espionage activities.

**Finding 11**<br>
Credential Theft (Finding 4) <br>
I-Soon likely focused on high-ranking individuals in government and technology sectors to bypass traditional cybersecurity defenses and gain privileged access to sensitive information. The hypothesis can be that the attackers were leveraging spear-phishing techniques or exploiting weak points in personal communications to steal login credentials. This stolen data would allow them to access privileged systems, confidential communications, and even critical decision-making processes. The ultimate goal of this credential theft could be to escalate permissions within an organization, leading to broader espionage activities across the network.

Additionally, the conversations in the files may hint at the reconnaissance phase of the credential theft. If we find discussions about accessing certain types of accounts or needing particular levels of permission (e.g., the mention of “intranet access” in File 1), these could point to targeted efforts to steal credentials necessary to penetrate deeper into the network.

**Question Form**<br>
Did I-Soon perform credential theft operations?

**Validation**<br>
The conversations, especially those about internal access control, suggest the possibility of discussing network permissions or credentials. The logs could reflect the reconnaissance or social engineering phase of a credential theft operation.
